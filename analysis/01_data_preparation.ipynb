```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
general_dam = pd.read_csv("dam_occupancy.csv")
individual_dam = pd.read_excel("istanbul-dams-daily-occupancy-rates.xlsx")
weather = pd.read_csv("weather_data.csv", skiprows=2)
```

```python
print("GENERAL DAM DATA")
display(general_dam.head())

print("INDIVIDUAL DAM DATA")
display(individual_dam.head())

print("WEATHER DATA")
display(weather.head())
```

```python
general_dam["DATE"] = pd.to_datetime(general_dam["DATE"])
general_dam = general_dam.set_index("DATE")

individual_dam["Tarih"] = pd.to_datetime(individual_dam["Tarih"])
individual_dam = individual_dam.set_index("Tarih")

weather["time"] = pd.to_datetime(weather["time"])
weather = weather.set_index("time")

print(general_dam.index)
print(individual_dam.index)
print(weather.index)
```

```python
dam_rate_monthly = general_dam["GENERAL_DAM_OCCUPANCY_RATE"].resample("ME").mean()
dam_water_monthly = general_dam["GENERAL_DAM_RESERVED_WATER"].resample("ME").mean()

dam_monthly = pd.concat([dam_rate_monthly, dam_water_monthly], axis=1)

display(dam_monthly.head())

rain_monthly = weather["precipitation_sum (mm)"].resample("ME").sum()
temp_monthly = weather["temperature_2m_mean (°C)"].resample("ME").mean()
rain_only_monthly = weather["rain_sum (mm)"].resample("ME").sum()
snow_monthly = weather["snowfall_sum (cm)"].resample("ME").sum()

weather_monthly = pd.concat([rain_monthly, temp_monthly, rain_only_monthly, snow_monthly],axis=1)

display(weather_monthly.head())

omerli_monthly = individual_dam["Ömerli"].resample("ME").mean()
darlik_monthly = individual_dam["Darlık"].resample("ME").mean()
elmali_monthly = individual_dam["Elmalı"].resample("ME").mean()
terkos_monthly = individual_dam["Terkos"].resample("ME").mean()
alibey_monthly = individual_dam["Alibey"].resample("ME").mean()
buyukcekmece_monthly = individual_dam["Büyükçekmece"].resample("ME").mean()
sazlidere_monthly = individual_dam["Sazlıdere"].resample("ME").mean()
kazandere_monthly = individual_dam["Kazandere"].resample("ME").mean()
pabucdere_monthly = individual_dam["Pabuçdere"].resample("ME").mean()
istrancalar_monthly = individual_dam["Istrancalar"].resample("ME").mean()

individual_monthly = pd.concat([
    omerli_monthly,
    darlik_monthly,
    elmali_monthly,
    terkos_monthly,
    alibey_monthly,
    buyukcekmece_monthly,
    sazlidere_monthly,
    kazandere_monthly,
    pabucdere_monthly,
    istrancalar_monthly
], axis=1)


display(individual_monthly.head())
```

```python
final_data = pd.concat([dam_monthly, weather_monthly], axis=1)

display(final_data.head())
print(final_data.columns)
```

```python
final_data = final_data.dropna()

display(final_data.head())
display(final_data.tail())
print(final_data.shape)
```
