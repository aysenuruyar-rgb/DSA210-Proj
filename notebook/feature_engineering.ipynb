```python
import pandas as pd
import numpy as np

df = pd.read_csv("final_dataset.csv", index_col=0)
df.index = pd.to_datetime(df.index)

df["dam_lag_1"] = df["GENERAL_DAM_OCCUPANCY_RATE"].shift(1)
df["dam_lag_3"] = df["GENERAL_DAM_OCCUPANCY_RATE"].shift(3)

df["rain_lag_1"] = df["precipitation_sum (mm)"].shift(1)
df["rain_lag_3"] = df["precipitation_sum (mm)"].shift(3)
df["snow_lag_1"] = df["snowfall_sum (cm)"].shift(1)
df["temp_lag_1"] = df["temperature_2m_mean (°C)"].shift(1)
df.head()
```

```python
df["rain_3m_avg"] = df["precipitation_sum (mm)"].rolling(3).mean()
df["temp_3m_avg"] = df["temperature_2m_mean (°C)"].rolling(3).mean()
df["dam_3m_avg"] = df["GENERAL_DAM_OCCUPANCY_RATE"].rolling(3).mean()

df["rain_snow_total"] = df["precipitation_sum (mm)"] + df["snowfall_sum (cm)"]

df["month"] = df.index.month
df["month_sin"] = np.sin(2 * np.pi * df["month"] / 12)
df["month_cos"] = np.cos(2 * np.pi * df["month"] / 12)

threshold = df["precipitation_sum (mm)"].median()
df["is_rainy"] = (df["precipitation_sum (mm)"] > threshold).astype(int)

df.head()
```

```python
df = df.dropna()
df.head()
```
